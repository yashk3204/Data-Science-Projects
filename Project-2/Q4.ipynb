```python
alpha0, beta0 = map(int, input("Enter alpha and beta values: ").split(','))
likes1, total1 = map(int, input("Enter likes and total for the first survey: ").split(','))
dislikes2, likes1 = map(int, input("Enter dislikes and likes for the second survey: ").split(','))

dislikes1 = total1 - likes1
alpha1 = alpha0 + likes1
beta1 = beta0 + dislikes1

likes2 = likes1 - dislikes2
alpha2 = alpha1 + likes2
beta2 = beta1 + dislikes2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
# Plotting
x = np.linspace(0, 1, 200)
plt.figure(figsize=(12, 8))
```

```python
# Prior beta distribution
prior_dist = beta(alpha0, beta0)
plt.plot(x, prior_dist.pdf(x), label="Prior Distribution", color='blue')

# Posterior distribution after first survey
posterior_dist_1 = beta(alpha1, beta1)
plt.plot(x, posterior_dist_1.pdf(x), label="Posterior after First Survey", color='green')

# Posterior distribution after second survey
posterior_second_dist = beta(alpha2, beta2)
plt.plot(x, posterior_second_dist.pdf(x), label="Posterior after Second Survey", color='red')

plt.title("Prior and Posterior Distributions")
plt.xlabel("p (Probability of liking the smartphone)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Explanation:
In Bayesian inference, we update the parameters of the Beta distribution by adding the number of "successes" (likes) and "failures" (dislikes) observed to the prior parameters.
In both cases, the posterior distribution of 𝑝 was derived by updating the parameters of the Beta distribution based on observed "successes" and "failures".
